```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('dogg.jpg',0)
edges = cv2.Canny(img,100,200)

plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
img = cv2.imread('Lenna.png',0)

edeges1 = cv2.Canny(img,50,200) 
edeges2 = cv2.Canny(img,100,200)
edeges3 = cv2.Canny(img,150,200)

plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(edeges1, cmap = 'gray')
plt.title('50,200'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(edeges2,cmap = 'gray')
plt.title('100,200'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(edeges3,cmap = 'gray')
plt.title('150,200'), plt.xticks([]), plt.yticks([])

plt.show()
```
